```python
#howmany passsanger are survived in dataset
survived_count = df['Survived'].sum()

# Display the number of survivors
print( survived_count)
```

```python
#howmany passsanger are survived in csv file were over 60 year of age
survived_over_60 = df[(df['Age'] > 60) & (df['Survived'] == 1)]

# Count the number of such passengers
survived_over_60_count = survived_over_60.shape[0]
print(survived_over_60_count)
```

```python
#what was the lawest and heighst fair paid to board  the titanic?store your ans in float
lowest_fare = df['Fare'].min()
highest_fare = df['Fare'].max()

# Store the results in a new DataFrame
fare_stats_df = pd.DataFrame({
    'Lowest_Fare': [float(lowest_fare)],
    'Highest_Fare': [float(highest_fare)]
})

# Save the results to a CSV file
fare_stats_df.to_csv('fare_statistics.csv', index=False)
print(fare_stats_df)
```
